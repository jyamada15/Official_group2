```python
import pandas as pd
import csv
import numpy as np
import time
from datetime import datetime
```

```python
path_1 = "./RawData_NYC Taxi2015/2015-01_100k.csv" #January 2015
path_2 = "./RawData_NYC Taxi2015/2015-02_100k.csv" #February 2015
path_3 = "./RawData_NYC Taxi2015/2015-03_100k.csv" #March 2015
path_4 = "./RawData_NYC Taxi2015/2015-04_100k.csv" #April 2015
path_5 = "./RawData_NYC Taxi2015/2015-05_100k.csv" #May 2015
path_6 = "./RawData_NYC Taxi2015/2015-06_100k.csv" #JUne 2015
path_7 = "./RawData_NYC Taxi2015/2015-07_100k.csv" #July 2015
path_8 = "./RawData_NYC Taxi2015/2015-08_100k.csv" #August 2015
path_9 = "./RawData_NYC Taxi2015/2015-09_100k.csv" #September 2015
path_10 ="./RawData_NYC Taxi2015/2015-10_100k.csv" #October 2015
path_11 ="./RawData_NYC Taxi2015/2015-11_100k.csv" #November 2015
path_12 ="./RawData_NYC Taxi2015/2015-12_100k.csv" #December 2015
```

```python
nyc_pd1 = pd.read_csv(path_1, encoding="UTF-8")
nyc_pd2 = pd.read_csv(path_2, encoding="UTF-8")
nyc_pd3 = pd.read_csv(path_3, encoding="UTF-8")
nyc_pd4 = pd.read_csv(path_4, encoding="UTF-8")
nyc_pd5 = pd.read_csv(path_5, encoding="UTF-8")
nyc_pd6 = pd.read_csv(path_6, encoding="UTF-8")
nyc_pd7 = pd.read_csv(path_7, encoding="UTF-8")
nyc_pd8 = pd.read_csv(path_8, encoding="UTF-8") 
nyc_pd9 = pd.read_csv(path_9, encoding="UTF-8") #September2015
nyc_pd10 = pd.read_csv(path_10, encoding="UTF-8") #October 2015
nyc_pd11 = pd.read_csv(path_11, encoding="UTF-8") #November 2015
nyc_pd12 = pd.read_csv(path_12, encoding="UTF-8") #Decemeber 2015

nyc_pd1.count() #
```

```python
#Consolidate all the NYC Taxi Trip data files separated by months into one(1) data set

nyc_app2 = nyc_pd1.append(nyc_pd2,ignore_index=True) #Jan, Feb 2015 appended

nyc_app3 = nyc_app2.append(nyc_pd3,ignore_index=True) #Jan,Feb, Mar 2015 appended

nyc_app4 = nyc_app3.append(nyc_pd4,ignore_index=True) #Jan,Feb, Mar, April 2015 appended

nyc_app5 = nyc_app4.append(nyc_pd5,ignore_index=True) ##Jan,Feb, Mar,April, May 2015 appended

nyc_app6Months = nyc_app5.append(nyc_pd6,ignore_index=True) #Jan,Feb, Mar,April, May, June 2015 appended

nyc_app7 = nyc_app6Months.append(nyc_pd7,ignore_index=True) #Jan,Feb, Mar,April, May, June, July 2015 appended

nyc_app8 = nyc_app7.append(nyc_pd8,ignore_index=True) #Jan,Feb, Mar,April, May, June,July, Aug 2015 appended

nyc_app9= nyc_app8.append(nyc_pd9,ignore_index=True) #Jan,Feb, Mar,April, May, June, July, Aug, Sept 2015 appended

nyc_app10= nyc_app9.append(nyc_pd10,ignore_index=True) #Jan,Feb, Mar,April, May, June, July, Aug, Sept, Oct 2015 appended

nyc_app11= nyc_app10.append(nyc_pd11,ignore_index=True) #Jan,Feb, Mar,April, May, June, July, Aug, Sept, Oct,Nov 2015 appended

nyc_app12Months= nyc_app11.append(nyc_pd12,ignore_index=True) #Jan,Feb, Mar,April,May,June,July, Aug, Sept, Oct,Nov,Dec 2015 appended

```

```python
#Convert the Compiled 12month data set into a series
nyc_Yr2015 = pd.DataFrame(nyc_app12Months)
nyc_Yr2015.head()
```
